```python

from TTS.api import TTS
from IPython.display import Audio, display, HTML

myLang = 'en-uk'
myTLD = 'co.uk'
#myText = 'The quick brown fox jumps over the lazy dog.'
#you need to create this file manually and insert the text of choice
myFile = open('/content/txt', 'r')
myText = myFile.read()
myFile.close()

myTTS = TTS(model_name="tts_models/en/ljspeech/tacotron2-DDC", progress_bar=False)

i = 0
paragraphs = myText.split('\n')
for paragraph in paragraphs:
  myText = paragraph
  print(myText)
  i+=1
  myFile = ('tts'+str(i)+'.wav')
  myTTS.tts_to_file(text=myText, file_path=myFile)
  display(HTML('<p>' + myTLD + myLang +'</p>'))
  display(Audio(myFile, autoplay=True))
  if i>3:
    break
```
